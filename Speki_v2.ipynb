```python
import tabula
```

```python
pdf_files = glob.glob('/content/speki_school/*.pdf')
pdf_files
```

```python
df_single = pd.DataFrame()
for pdf_file in pdf_files:
    pdf_tables = tabula.read_pdf(pdf_file,
                             pages='30-50',
                             multiple_tables=True,
                             lattice=True)
    for pdf_table in pdf_tables:
        col_count = len(pdf_table.columns)
        if (col_count >= 3 and col_count <= 11) and pdf_table.columns[0] == 'Unnamed: 0':
            df_single = pd.concat([df_single, pdf_table])
        else:
            continue
```
